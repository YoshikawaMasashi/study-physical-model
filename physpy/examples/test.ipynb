```python
import os

import matplotlib.pyplot as plt
import numpy as np

import physpy
```

```python
os.makedirs("./results", exist_ok=True)
```

```python
fdm = physpy.physpy.FDMRunner((300, 300), 1.0, 0.1, 0.1, (100, 100))
ret = fdm.run(3000)
```

```python
for i, ret_ in enumerate(ret[::10]):
    plt.clf()
    plt.imshow(ret_, vmin=-5, vmax=5, cmap="seismic")
    plt.savefig(f"./results/{i}.png")
```
